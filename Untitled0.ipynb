```python
# Import necessary libraries
from transformers import pipeline
import matplotlib.pyplot as plt

# Define a class for emotion detection
class EmotionDetector:
    def __init__(self, model_name="j-hartmann/emotion-english-distilroberta-base"):
        """
        Initialize the emotion detection model.

        Parameters:
            model_name (str): Name of the Hugging Face model for emotion detection.
        """
        try:
            # Load the emotion classification pipeline with the specified model
            self.emotion_pipeline = pipeline("text-classification", model=model_name, return_all_scores=True)
            print("Model loaded successfully.")
        except Exception as e:
            print("Error loading model:", e)

    def analyze_emotion(self, text):
        """
        Analyze emotions of a given text.

        Parameters:
            text (str): The text to analyze.

        Returns:
            dict: Dictionary with emotions and their scores.
        """
        try:
            results = self.emotion_pipeline(text)
            # Sort emotions by score in descending order
            sorted_results = sorted(results[0], key=lambda x: x['score'], reverse=True)
            return {emotion['label']: emotion['score'] for emotion in sorted_results}
        except Exception as e:
            print("Error analyzing emotion:", e)
            return {}

    def plot_emotions(self, emotions):
        """
        Plot the emotions with their respective scores.

        Parameters:
            emotions (dict): Dictionary with emotions and their scores.
        """
        try:
            # Unpack the emotion labels and scores for plotting
            labels = list(emotions.keys())
            scores = list(emotions.values())

            # Plot using matplotlib
            plt.figure(figsize=(10, 5))
            plt.barh(labels, scores, color='skyblue')
            plt.xlabel("Score")
            plt.title("Emotion Detection Results")
            plt.gca().invert_yaxis()  # Invert y-axis to show highest score on top
            plt.show()
        except Exception as e:
            print("Error plotting emotions:", e)

# Function to analyze multiple texts
def analyze_multiple_texts(texts, model_name="j-hartmann/emotion-english-distilroberta-base"):
    """
    Analyze a list of texts for emotions and display results.

    Parameters:
        texts (list): List of texts to analyze.
        model_name (str): Name of the Hugging Face model for emotion detection.
    """
    detector = EmotionDetector(model_name=model_name)
    for text in texts:
        print(f"\nAnalyzing text: {text}")
        emotions = detector.analyze_emotion(text)
        print("Detected emotions and scores:", emotions)
        detector.plot_emotions(emotions)

# Example usage with sample texts
texts = [
    "I am feeling so happy and excited today!",
    "This is really frustrating and makes me angry.",
    "I love to make friends.",
    "I’m worried about my future.",
    "I just feel so tired and drained of energy."
]

# Run the emotion analysis for each text
analyze_multiple_texts(texts)
```
